# **SOAL PENJABARAN 2**

> Apa saja jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016?




```python
Data2=client.query('''
SELECT DISTINCT primary_type
FROM `bigquery-public-data.austin_crime.crime`
where EXTRACT(YEAR FROM timestamp) = 2016

''').to_dataframe()
Data2
```

# **SOAL PENJABARAN 3**
Berapa banyak kasus tindak kejahatan berdasarkan kasus pembersihannya (clearance status) selama tahun 2016?

```python
Crime = client.query('''
SELECT  clearance_status,
        COUNT(clearance_status) AS Jumlah
FROM `bigquery-public-data.austin_crime.crime`
where EXTRACT(YEAR FROM timestamp) = 2016
GROUP BY clearance_status
ORDER BY Jumlah DESC

''').to_dataframe()
Crime
```

Kode di atas menunjukan total kejahatan yang terjadi berdasarkan jenis penyelesaiannya. Dapat dilihat pada tabel di atas bahwa kasus yang belum diselesaikan memiliki jumlah terbanyak dengan total 29715 kasus dan diikuti oleh cleared by arrest sebesar 5013 dan cleared by exxception sebesar 1014. Hal ini menunjukan diperlukannya peningkatan kualitas dari penegak hukum dalam menyelesaikan masalah karena kasus yang belum selesai memiliki angka yang sangat jauh dibandingkan yg lainnya.

# **SOAL PENJABARAN 4**

> Jenis kejahatan apa saja yang paling banyak belum terselesaikan kasusnya di tahun 2016?




```python
CrimeNotFinished = client.query('''
SELECT  primary_type,
        COUNT(primary_type) AS Jumlah
FROM `bigquery-public-data.austin_crime.crime`
where EXTRACT(YEAR FROM timestamp) = 2016 AND
clearance_status = 'Not cleared'
GROUP BY primary_type
ORDER BY Jumlah DESC

''').to_dataframe()
CrimeNotFinished
```

Pada tabel diatas menunjukan jenis kasus dan total dari kasus tersebut terjadi  yang belum terselesaikan selama tahun 2016. Kasus theft/pencurian menjadi kasus yang paling banyak belum terselesaikan di tahun 2016. Dengan data ini kita dapat menentukan strategi untuk mengurangi kasus kejahatan berupa pencurian yang terjadi di tahun2 berikutnya. Salah satu strategi yang mungkin bisa dilakukan adalah dengan menambah personil kepolisian untuk melakukan patroli di sepanjang jalan khususnya saat malam hari.

# **SOAL PENJABARAN 5**

> Di bulan apa di tahun 2016 kasus pencurian (theft) sangat banyak terjadi?




```python
CrimeNotFinished = client.query('''
SELECT  primary_type as Jenis_Kriminal,
        COUNT(primary_type) AS Jumlah,
        EXTRACT(month FROM timestamp) as month
FROM `bigquery-public-data.austin_crime.crime`
where EXTRACT(YEAR FROM timestamp) = 2016 AND
primary_type = 'Theft'
GROUP BY Jenis_Kriminal,month
ORDER BY Jumlah DESC

''').to_dataframe()
CrimeNotFinished
```

# **SOAL PENJABARAN 6**


> Di distrik apa yang paling banyak terjadi kasus pencurian selama 2016?






```python
District=client.query('''

SELECT district,
      Count(district) as jumlah_kasus
FROM `bigquery-public-data.austin_crime.crime`
WHERE
      EXTRACT(YEAR FROM timestamp) = 2016 AND
      primary_type = 'Theft'
GROUP BY
      district
ORDER BY
      jumlah_kasus DESC

''').to_dataframe()
District
```

Tabel di atas menunjukan distrik mana yang memiliki kasus pencurian terbanyak sepanjang tahun 2016. Data ini dapat membantu kepolisian untuk membuat strategi patroli yang tepat untuk mengurangi kasus pencurian tersebut.
Salah satu strategi yang dapat dilakukan mungkin dengan cara mengurangi personil yang berpatroli di daerah yang memiliki jumlah kasus terrendah dan memindahkan personil polisi tersebut untuk berpatroli di daerah yang memiliki kasus tertinggi.

# **SOAL PENJABARAN 7**


> Berapa lama rata-rata waktu (dalam hari) kasus tindak kejahatan 'theft' terselesaikan sejak kasus tersebut terangkat selama tahun 2016? (Hint: Hitung selisih hari dari timestamp hingga clearance date)



```python
Rata_waktu =client.query('''
SELECT primary_type,
       AVG(TIMESTAMP_DIFF(clearance_date, timestamp, DAY)) AS Rata_Rata
FROM `bigquery-public-data.austin_crime.crime`
WHERE primary_type = 'Theft'
  AND clearance_status LIKE 'Cleared%'
  AND EXTRACT(YEAR FROM timestamp) = 2016
GROUP BY primary_type


''').to_dataframe()
Rata_waktu
```

Berdasarkan data di atas dapat dilihat kasus pencurian memiliki rata2 waktu selama 14 hari untuk terselesaikan. Hal tersebut dapat dianalisis lebih jauh lagi apakah waktu tersebut tergolong lama atau tidak. Jika memang waktu tersebut dianggap terlalu lama maka dapat mencari strategi baru untuk pembenahan agar waktu penyelesaian akan lebih cepat. Jika memang waktu tersebut dirasa sudah pantas berarti kepolisian sudah melakukan kinerja yang baik .
